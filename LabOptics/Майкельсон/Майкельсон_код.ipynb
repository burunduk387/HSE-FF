```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([5, 10, 15, 20, 25, 30])
N = np.array([19, 40, 57, 75, 90, 110])
z = np.polyfit(x, N, 1)
z = np.poly1d(z)
xx = np.linspace(4, 31, 180)
plt.plot(xx, z(xx), label="Поддогнанная кривая", color="orange")
plt.scatter(x, N, label="Экспериментальные данные")
plt.xlabel(r"$\delta x$, мкм")
plt.ylabel(r"$N$, шт.")
plt.title("Зависимость числа переходов от смещения зеркала")
plt.grid()
plt.legend()
plt.savefig("maic1.png")
plt.tight_layout()
plt.show()
print(z)
print(np.mean(2*x/N))
print(np.std(2*x/N))
```

```python
def n(N, l, t, a):
    return ((N*l/(2*t) + np.cos(np.radians(a)) - 1)**2 + (np.sin(np.radians(a)))**2)/(2 * (- N*l/(2*t) - np.cos(np.radians(a)) + 1))
```

```python
a = np.array([ 2, 3, 4, 5, 6])
N = np.array([7, 16, 25, 38, 49])
print(n(N, 532, 7500000, a))
print(np.mean(n(N, 532, 7500000, a)))
print(np.std(n(N, 532, 7500000, a)))
```

```python
a = np.array([1, 1.5, 2, 2.5, 3, 3.5, 4])
N = np.array([2, 5, 9, 13, 19, 25, 29])
print(n(N, 532, 12000000, a))
print(np.mean(n(N, 532, 12000000, a)))
print(np.std(n(N, 532, 12000000, a)))
```

```python
x = np.array([25, 27, 30, 35, 40, 45, 50, 55, 60])
N = np.array([0, 1, 3, 5, 7, 10, 12, 15, 17])
z = np.polyfit(x, N, 1)
z = np.poly1d(z)
xx = np.linspace(25, 61, 180)
plt.plot(xx, z(xx), label="Поддогнанная кривая", color="orange")
plt.scatter(x, N, label="Экспериментальные данные")
plt.xlabel(r"$T, ^\circ C$")
plt.ylabel(r"$N$, шт.")
plt.title("Зависимость числа переходов от температуры")
plt.grid()
plt.legend()
plt.savefig("maic2.png")
plt.tight_layout()
plt.show()
print(z)
dT = np.zeros(x.shape[0] - 1)
dN = np.zeros(x.shape[0] - 1)
for i in range(1, x.shape[0]):
    dT[i - 1] = x[i] - x[0]
    dN[i - 1] = N[i] - N[0]
print(dT, dN)
print(dN*532/(9e6*dT))
print(np.mean(dN*532/(9e6*dT)))
print(np.std(dN*532/(9e6*dT)))
```
