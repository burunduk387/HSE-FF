```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def angle(a, c):
    t = a/c
    return np.arctan(t)
```

```python
plt.figure(figsize=(10, 7))
l = 130
a = 0
c = 51.5
x = [118, 141, 161, 181, 202]
angle0 = angle(a, c)
dx = np.zeros(5)
for i in range(len(x)):
    dx[i] = x[i] - x[2]
maxi = np.arange(-2, 3)
L = np.array([l, l, l, l, l])
z = np.polyfit(maxi, np.sin(angle0) + np.sin(angle(dx, L)), 1)
z = np.poly1d(z)
xx = np.linspace(-2.01, 2.01, 10)
plt.grid()
plt.plot(xx, z(xx), label="Fitted line", color="orange")
plt.title(r"$\sin{\theta} + \sin{\theta_i}$ as f($m$)", fontsize=16)
plt.xlabel("m", fontsize=16)
plt.ylabel(r"$\sin{\theta} + \sin{\theta_i}$", fontsize=16)
plt.scatter(maxi, np.sin(angle0) + np.sin(angle(dx, L)), label="Experimental data")

z = np.polyfit(maxi, np.sin(angle0) + np.sin(angle(dx, L)), 1)
k = z[0]/520*1e6


a = 25
c = 51.5
x = [49, 80.5, 105.5, 127, 147]
angle0 = angle(a, c)
print(angle0)
dx = np.zeros(5)
for i in range(len(x)):
    dx[i] = x[i] - x[2]
maxi = np.arange(-2, 3)
L = np.array([l, l, l, l, l])
z = np.polyfit(maxi, np.sin(angle0) + np.sin(angle(dx, L)), 1)
z = np.poly1d(z)
plt.plot(xx, z(xx), label=r"Fitted line, $\theta_i = 0.45 rad$", color="pink")
plt.scatter(maxi, np.sin(angle0) + np.sin(angle(dx, L)), label=r"Experimental data, $\theta_i = 0.45 rad$", color="black")
z = np.polyfit(maxi, np.sin(angle0) + np.sin(angle(dx, L)), 1)
k1 = z[0]/520*1e6

print(np.mean([k, k1]))

plt.legend()
plt.tight_layout()
plt.savefig("300.png")
```
