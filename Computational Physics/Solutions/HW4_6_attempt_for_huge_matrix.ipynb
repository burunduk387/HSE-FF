```python
import numpy as np
from scipy.sparse.linalg import eigs, eigsh
import matplotlib.pyplot as plt
```

```python
n = 3
u = np.random.normal(0, 1, n)
d = np.random.normal(0, 1, n)
u = u[:, np.newaxis]
s = 1 / np.sqrt(u.T @ u)
#Golub
#Eigenvalues of a matrix modified by a rank one matrix
print(np.min(np.linalg.eigvalsh(np.diag(d)+u@u.T*s)))
#mm = np.min(np.linalg.eigvals(B))
#print(mm)
#evals, evecs = eigsh(B, 1, which = 'SM')
#print (evals)
y = []
x = np.linspace(-5, 5, 2000)
for i in x:
    k = 0
    for j in range(len(u)):
        k += s[0, 0] * (u[j]**2) / (d[j] - i)
    k += 1
    y.append(k)
plt.plot(x, y)

```
