```python
# Ax = b -> x = Bx + c
# B = E - D^(-1) * A
# B = D^(-1) * (D - A), c = D^(-1) * b
import numpy as np
```

```python
rnd = np.random.RandomState(1234)
n = 10
A = rnd.uniform(size=(n, n)) + np.diag([15]*n)
b = rnd.uniform(size=n)
D = np.diag(np.diag(A))
D1 = np.linalg.inv(D) 
B = D1 @ (D - A)
c = D1 @ b
x = np.random.random(n)
xp = np.random.random(n)
q = np.linalg.norm(B)
print("Норма:", q)

#stop condition
eps = 1.0E-18
stop = (1 - q)/q * eps

i = 0

while np.linalg.norm(x - xp) > stop:
    i += 1
    xp = x
    x = B @ xp + c
print("ITER NUM:", i)   
print("Решение ли:", x - (B@x + c))


# Now check for changed diagonal A. Apparently if norm(B) >= 1 method should not work
for i in range(n):
    A[i, i] *= (2/3)
D = np.diag(np.diag(A))
D1 = np.linalg.inv(D) 
B = D1 @ (D - A)
c = D1 @ b
x = np.random.random(n)
xp = np.random.random(n)
q = np.linalg.norm(B)
print("Норма:", q)

#stop condition
eps = 1.0E-18
stop = (1 - q)/q * eps

i = 0

while np.linalg.norm(x - xp) > stop:
    i += 1
    xp = x
    x = B @ xp + c
print("ITER NUM:", i)
print("Решение ли:", x - (B@x + c))

# Now diag(A) / 3
for i in range(n):
    A[i, i] /= 2
D = np.diag(np.diag(A))
D1 = np.linalg.inv(D) 
B = D1 @ (D - A)
c = D1 @ b
x = np.random.random(n)
xp = np.random.random(n)
q = np.linalg.norm(B)
print("Норма:", q)

# Stop condition, changed to bigger eps
eps = 1.0E-2
stop = (1 - q)/q * eps

# Handbrake
hb = 1.0E3
i = 0

while np.linalg.norm(x - xp) > stop and i < hb:
    i += 1
    xp = x
    x = B @ xp + c
if i == hb:
    print("Method not working")
else:
    print("ITER NUM:", i)
```
