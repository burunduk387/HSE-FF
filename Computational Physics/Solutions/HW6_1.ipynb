```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#consts for i and ii
n = 10
x0 = 1
```

```python
#i
x = np.zeros(n)
x[0] = x0
for i in range(1, n):
    x[i] = x[i - 1] + (np.cos(x[i - 1]) + 1) / np.sin(x[i - 1])   
erf = [(x[i]-np.pi)**2 for i in range(n)]

print("Линейная сходимость")
plt.yscale('log')
plt.plot(erf) 
plt.grid() 
plt.show()

#ii
x = np.zeros(n)
x[0] = x0
for i in range(1, n):
     x[i] = x[i - 1] / 2 + 1 / x[i - 1]
erf = [(x[i]-np.sqrt(2))**2 for i in range(n)]
print("Квадратичная сходимость")
plt.yscale('log')
plt.plot(erf) 
plt.grid() 
plt.show()
```
