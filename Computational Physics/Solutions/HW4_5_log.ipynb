```python
def kroneker (i, j):
    if i == j:
        return 1
    return 0
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg
n = 32
A = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        A[i, j] = -kroneker(i, j) + kroneker(i, j-1) + kroneker (i, j-2)
#Вообще из вида А очевидно, что её собственные значения на главной диагонали, и все, в силу определения матрицы, -1
l = np.linalg.eigvals(A)
t = np.linspace(0, 50, 1000)
y = []
for i in t:
    y.append(np.linalg.norm(scipy.linalg.expm(A*i), 2))
plt.plot(t, y)
```
